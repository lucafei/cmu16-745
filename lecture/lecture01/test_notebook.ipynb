```julia
# 简单的摆动力学（来自你的 pendulum.jl）
function pendulum_dynamics(θ, ω, u)
    m = 1.0  # 质量
    l = 1.0  # 长度
    g = 9.81 # 重力加速度
    
    θ̇ = ω
    ω̇ = -g/l * sin(θ) + u/(m*l^2)
    
    return θ̇, ω̇
end

# 测试
θ, ω = pendulum_dynamics(π/4, 0.0, 0.0)
println("θ̇ = ", θ, ", ��̇ = ", ω)
```
